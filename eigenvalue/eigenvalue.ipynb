### Eigen mean latent or characterstics
Equation: x<sub>k+1</sub> = Ax<sub>k</sub>

Equation of Eigen: Ax = kx

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# matrix A 
A = np.array([[3, -1], [-1,3]])

# x vector (eigen vector)
x = np.array([[1], [-1]])

# caluculating the product A@x
result = A@x

# using dark back ground for the plot
plt.style.use('dark_background')  # Enable dark mode

# Plot x (deepskyblue) and result (orange)
plt.quiver(0, 0, x[0,0], x[1,0], angles='xy', scale=1, scale_units='xy', color='deepskyblue', label='x', width=0.005)
plt.quiver(0, 0, result[0,0], result[1,0], angles='xy', scale_units='xy', scale=1, color='orange', label='A@x', width=0.005)

plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.grid(True, color='gray', linestyle='--', linewidth=0.5, alpha=0.9)

plt.title('Vectors x and A@x')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### How do I find these special vectors and eigen values

solve det(A - kI) = 0

```python
eigenvalues, eigenvectors = np.linalg.eig(A)  
eigenvalues
```
